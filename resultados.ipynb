```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Se lee el fichero resultados.csv y se almacena en un dataframe

```python
df = pd.read_csv('resultados.csv', sep=';')
df
```

Se genera una grafica de número de vertices vs tiempo requerido para hayar la solución

```python
# Crear el gráfico de num_colors vs elapsed_time
plt.figure(figsize=(10, 6))
plt.scatter(df['vertices'], df['tiempo'], color='red', marker='o')

# Agregar etiquetas para los nombres de los puntos
for i, row in df.iterrows():
    plt.annotate(row['nombre'], (row['vertices'], row['tiempo']), textcoords="offset points", xytext=(2,2), ha='left', rotation=50)

plt.xlabel('Vertices')
plt.ylabel('Tiempo(s)')
plt.title('Vertices vs. Tiempo(s)')
plt.grid(True)
plt.show()
```
